````python
def compare_values(column_name, dataframe):
    # ë°ì´í„°í”„ë ˆì„ì˜ íŠ¹ì • ì»¬ëŸ¼ì„ ê°€ì ¸ì˜¤ê¸°
    column_values = dataframe[column_name]

    # ë°ì´í„°í”„ë ˆì„ì˜ íŠ¹ì • ì»¬ëŸ¼ì˜ í¬ê¸° ë¹„êµ ì¶œë ¥
    for i in range(len(column_values) - 1):
        current_value = column_values.iloc[i]
        next_value = column_values.iloc[i + 1]

        if current_value < next_value:
            print("-")
        elif current_value > next_value:
            print('+')
        else:
            print('=')
````

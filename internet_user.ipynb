```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python


# ì›¹í˜ì´ì§€ì˜ URL ì§€ì •
url = 'https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population'

# ì›¹í˜ì´ì§€ì˜ HTMLì„ ê°€ì ¸ì˜¤ê¸°
response = requests.get(url)
html = response.text

# BeautifulSoupì„ ì‚¬ìš©í•˜ì—¬ HTML íŒŒì‹±
soup = BeautifulSoup(html, 'html.parser')

# HTMLì—ì„œ í…Œì´ë¸” ìš”ì†Œë¥¼ ì°¾ê¸°
tables = soup.find_all('table',class_="wikitable sortable sticky-header col2left col6left")
```

```python
# í…Œì´ë¸”ë“¤ì„ DataFrameìœ¼ë¡œ ë³€í™˜
dfs = []
for table in tables:
    df = pd.read_html(str(table))[0]
    dfs.append(df)
```

```python
# ì—¬ëŸ¬ í…Œì´ë¸”ì´ ìˆë‹¤ë©´ í•„ìš”ì— ë”°ë¼ ì ì ˆí•œ ì²˜ë¦¬ë¥¼ ìˆ˜í–‰í•˜ê³  í•©ì¹  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
df_population = pd.concat(dfs, axis=0, ignore_index=True)
df_population
```

```python
def income_average(group):
    return np.average(group['income_per_person'], weights=group['population'])

def internet_average(group):
    return np.average(group['internet_use_rate'], weights=group['population'])


grouped = df.groupby(['region', 'sub_region'])

income_avg = grouped.apply(income_average).reset_index(name='weighted_ave_income')
internet_avg = grouped.apply(internet_average).reset_index(name='weighted_ave_internet')

result = internet_avg.set_index(['region', 'sub_region']).join(income_avg.set_index(['region', 'sub_region']))
```

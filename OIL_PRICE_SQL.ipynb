````python
gmaps_key = 'AIzaSyDS8JfW9b9Fc_DasLH4INDxCoy8aU79UfA'
gmaps = googlemaps.Client(key=gmaps_key)
````

````python
# 화폐단위 문자형을 입력받아 숫자형으로 반환하는 함수 (테스트 입력 : ‘1,000’)
def remove_commas(number_str):
    # 숫자 문자열에서 ',' 제거
    result = number_str.replace(',', '')
    return int(result)

remove_commas('1,000')
````

````python
#주유소 브랜드를 입력하면 GAS_BRAND 데이터를 참고하여 ID 를 반환하는 함수 (테스트 입력 : ‘SK에너지’)
def brand_search(n):
    if n == 'SK에너지':
        return 1

    elif n == 'HD현대오일뱅크':
        return 2

    elif n == 'GS칼텍스':
        return 3

    elif n == 'S-OIL':
        return 4

    elif n == '알뜰주유소':
        return 5

    else:
        return 6

brand_search('SK에너지')
````

````python
# 주소를 입력받아 구 이름을 반환하는 함수 (테스트 입력 : ‘서울시 강남구 헌릉로 730’)
def split_adr(n):
    tmp  = re.split(' ', n)
    return tmp[1]
split_adr('서울시 강남구 헌릉로 730')
````

````python
# 주소를 입력받아 위도, 경도를 반환하는 함수 (테스트 입력 : ‘서울시 강남구 헌릉로 730’)
def get_lat_lng(n):
    tmp = gmaps.geocode(n, language='ko')
    lat = tmp[0].get('geometry')['location']['lat']
    lat = round(lat, 7)
    lng = tmp[0].get('geometry')['location']['lng']
    lng = round(lng, 7)
    return lat, lng

get_lat_lng('서울시 강남구 헌릉로 730')
````
